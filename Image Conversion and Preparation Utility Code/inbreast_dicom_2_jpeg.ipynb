```python
import os
import numpy as np
!pip install pydicom pymedphys
import pydicom
from PIL import Image

from google.colab import drive
drive.mount('/content/gdrive')

def get_names_of_imgs_inside_folder(directory):

    names = []

    for root, dirnames, filenames in os.walk(directory):
        for filename in filenames:
            _, ext = os.path.splitext(filename)
            if ext in [".dcm"]:
                names.append(filename)

    return names

def convert(directory):
    im = pydicom.dcmread('/content/gdrive/MyDrive/polygence/datasets/inbreast/AllDICOMs/'+directory)
    im = im.pixel_array.astype(float)

    rescaled_image = (np.maximum(im,0)/im.max())*255 # float pixels
    final_image = np.uint8(rescaled_image) # integers pixels

    final_image = Image.fromarray(final_image)
    return final_image

names = get_names_of_imgs_inside_folder('/content/gdrive/MyDrive/polygence/datasets/inbreast/AllDICOMs')

for name in names:
    image = convert(name)
    outputfile = "/content/gdrive/MyDrive/polygence/datasets/inbreast/img1"+".jpeg"
    out_file = os.path.basename(name.split('.')[0])
    output_full_file_path= "/content/gdrive/MyDrive/polygence/datasets/inbreast/img/"+out_file+".jpeg"
    #image.show()
    #image.save(f"/content/gdrive/MyDrive/polygence/datasets/inbreast/img/rahul.jpg")
    image.save(output_full_file_path)
```
